```python
# Create a function to test the OpenAI API
# Import the OpenAI package
from openai import OpenAI 

def simple_test():
  client = OpenAI(api_key=openai.api_key)

  completion = client.chat.completions.create(
  model="gpt-4o-mini",
  messages=[
    {"role": "user", "content": "Explain in three sentences or less why the sky appears blue"}
  ]
)

  print(completion.choices[0].message.content)
```

```python
# Test simple_test()
simple_test() 
```
