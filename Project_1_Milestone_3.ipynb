```python
# Create some sample documents
sample_documents = {
    "business_contract": """Service Agreement
    This agreement is between TechCorp Inc. and Global Solutions LLC.
    Effective Date: January 15, 2024
    Contact: John Smith, CEO of TechCorp
    Location: Seattle, Washington
    This contract establishes the terms for software development services.""",

    "technical_report": """System Performance Analysis Report
    Prepared by: Dr. Sarah Chen, Senior Engineer
    Date: March 10, 2024
    Location: San Francisco Data Center
    The database performance improved by 40% after optimization.
    Microsoft Azure infrastructure was utilized for testing.""",

    "informal_email": """Hey team,
    Quick update on the project. Can we meet tomorrow at 2pm in New York?
    The client Microsoft wants to discuss the timeline.
    Thanks, Mike Johnson
    Sent: February 5, 2024""",

    "project_proposal": """Project Proposal: AI Integration
    Submitted by: DataTech Solutions
    Client: Amazon Web Services
    Proposal Date: April 1, 2024
    Project Manager: Lisa Rodriguez
    Location: Austin, Texas
    This formal proposal outlines our AI implementation strategy.""",

    "meeting_minutes": """Weekly Team Meeting - March 15, 2024
    Attendees: Bob Wilson, Kate Miller, Tom Davis
    Location: Chicago Office
    Topics discussed: Q2 budget planning with IBM partnership
    Next meeting: March 22, 2024"""
}
```

```python
# Ground truth for comaprison
ground_truth = {
    "business_contract": {
        "entities": {
            "People": ["John Smith"],
            "Organizations": ["TechCorp Inc.", "Global Solutions LLC"],
            "Locations": ["Seattle", "Washington"],
            "Dates": ["January 15, 2024"]
        },
        "classification": {"type": "business", "formality": "formal", "purpose": "contract"}
    },
    "technical_report": {
        "entities": {
            "People": ["Dr. Sarah Chen"],
            "Organizations": ["Microsoft Azure"],
            "Locations": ["San Francisco Data Center"],
            "Dates": ["March 10, 2024"]
        },
        "classification": {"type": "technical", "formality": "formal", "purpose": "report"}
    },
    "informal_email": {
        "entities": {
            "People": ["Mike Johnson"],
            "Organizations": ["Microsoft"],
            "Locations": ["New York"],
            "Dates": ["February 5, 2024"]
        },
        "classification": {"type": "business", "formality": "informal", "purpose": "email"}
    },
    "project_proposal": {
        "entities": {
            "People": ["Lisa Rodriguez"],
            "Organizations": ["DataTech Solutions", "Amazon Web Services"],
            "Locations": ["Austin", "Texas"],
            "Dates": ["April 1, 2024"]
        },
        "classification": {"type": "business", "formality": "formal", "purpose": "proposal"}
    },
    "meeting_minutes": {
        "entities": {
            "People": ["Bob Wilson", "Kate Miller", "Tom Davis"],
            "Organizations": ["IBM"],
            "Locations": ["Chicago Office"],
            "Dates": ["March 15, 2024", "March 22, 2024"]
        },
        "classification": {"type": "business", "formality": "formal", "purpose": "other"}
    }
}
```

```python
# Initialize analyzer and results storage
analyzer = DocumentAnalyzer()
test_results = {}
processing_times = {}

print("=== DOCUMENT ANALYSIS TESTING ===\n")
```

```python
# Run analysis on all sample documents
for doc_name, doc_text in sample_documents.items():
    print(f"Testing {doc_name}...")
    start_time = time.time()

    try:
        # Run analysis
        entities = analyzer.extract_entities(doc_text)
        summary = analyzer.generate_summary(doc_text)
        classification = analyzer.classify_document(doc_text)

        test_results[doc_name] = {
            "entities": entities,
            "summary": summary,
            "classification": classification
        }

        processing_times[doc_name] = time.time() - start_time
        print(f"{doc_name} processed in {processing_times[doc_name]:.2f} seconds\n")

    except Exception as e:
        print(f"Error processing {doc_name}: {e}\n")
        test_results[doc_name] = None

```

```python
# Comparison functions
def compare_entities(ai_entities, manual_entities):
    """Compare AI entity extraction with manual ground truth."""
    if not ai_entities or not manual_entities:
        return 0

    total_correct = 0
    total_manual = 0

    for category in manual_entities:
        manual_list = manual_entities.get(category, [])
        ai_list = ai_entities.get(category, [])

        total_manual += len(manual_list)

        for manual_entity in manual_list:
            if any(manual_entity.lower() in ai_entity.lower() or ai_entity.lower() in manual_entity.lower()
                   for ai_entity in ai_list):
                total_correct += 1

    return (total_correct / total_manual * 100) if total_manual > 0 else 0

def compare_classification(ai_class, manual_class):
    """Compare AI classification with manual ground truth."""
    if not ai_class or not manual_class:
        return 0

    correct = 0
    total = len(manual_class)

    for key in manual_class:
        if ai_class.get(key, "").lower() == manual_class[key].lower():
            correct += 1

    return (correct / total * 100) if total > 0 else 0

```

```python
# Calculate accuracy for each document
accuracy_results = {}
print("=== ACCURACY ANALYSIS ===\n")

for doc_name in sample_documents.keys():
    if test_results[doc_name] is None:
        continue

    entity_accuracy = compare_entities(
        test_results[doc_name]["entities"],
        ground_truth[doc_name]["entities"]
    )

    classification_accuracy = compare_classification(
        test_results[doc_name]["classification"],
        ground_truth[doc_name]["classification"]
    )

    accuracy_results[doc_name] = {
        "entity_accuracy": entity_accuracy,
        "classification_accuracy": classification_accuracy
    }

    print(f"{doc_name}:")
    print(f"  Entity Extraction: {entity_accuracy:.1f}%")
    print(f"  Classification: {classification_accuracy:.1f}%")
    print()

```

```python
# Edge case testing
print("=== EDGE CASE TESTING ===\n")

edge_cases = {
    "very_short": "Meeting tomorrow at 3pm.",
    "very_long": "This is a very long document. " * 1000,
    "empty": "",
    "special_chars": "Report by João García from São Paulo on 15/03/2024.",
    "numbers_only": "123 456 789 2024",
    "mixed_languages": "Meeting with François from Paris about the project."
}

edge_case_results = {}
for case_name, case_text in edge_cases.items():
    print(f"Testing edge case: {case_name}")
    try:
        entities = analyzer.extract_entities(case_text)
        classification = analyzer.classify_document(case_text)
        edge_case_results[case_name] = "Success"
        print(f"  Processed successfully")
    except Exception as e:
        edge_case_results[case_name] = f"Error: {str(e)}"
        print(f"  Error: {e}")
    print()

```

```python
# Final validation report
print("=== VALIDATION REPORT ===\n")

# Calculate overall metrics
if accuracy_results:
    avg_entity_accuracy = sum(r["entity_accuracy"] for r in accuracy_results.values()) / len(accuracy_results)
    avg_classification_accuracy = sum(r["classification_accuracy"] for r in accuracy_results.values()) / len(accuracy_results)
    avg_processing_time = sum(processing_times.values()) / len(processing_times)

    print(f"Overall Performance:")
    print(f"  Average Entity Extraction Accuracy: {avg_entity_accuracy:.1f}%")
    print(f"  Average Classification Accuracy: {avg_classification_accuracy:.1f}%")
    print(f"  Average Processing Time: {avg_processing_time:.2f} seconds")
    print(f"  Edge Cases Handled: {sum(1 for r in edge_case_results.values() if r == 'Success')}/{len(edge_cases)}")
    print()

    # Identify best/worst performing document types
    best_entity = max(accuracy_results.items(), key=lambda x: x[1]["entity_accuracy"])
    worst_entity = min(accuracy_results.items(), key=lambda x: x[1]["entity_accuracy"])

    print(f"Best Entity Extraction: {best_entity[0]} ({best_entity[1]['entity_accuracy']:.1f}%)")
    print(f"Worst Entity Extraction: {worst_entity[0]} ({worst_entity[1]['entity_accuracy']:.1f}%)")
    print()

    print("Deployment Recommendations:")
    if avg_entity_accuracy >= 80 and avg_classification_accuracy >= 80:
        print("  System ready for deployment with high accuracy")
    elif avg_entity_accuracy >= 70 and avg_classification_accuracy >= 70:
        print("  System suitable for deployment with human review")
    else:
        print("  System needs improvement before deployment")

    print(f"  Estimated cost per document: ~$0.01-0.03 based on processing time")
    print(f"  Recommended for: {', '.join([doc for doc, acc in accuracy_results.items() if acc['entity_accuracy'] >= 75])}")

# Save detailed results
with open("validation_results.json", "w", encoding="utf-8") as f:
    json.dump({
        "test_results": test_results,
        "accuracy_results": accuracy_results,
        "processing_times": processing_times,
        "edge_case_results": edge_case_results
    }, f, indent=2)

print("\nDetailed results saved to validation_results.json")
print("=== TESTING COMPLETE ===")
```
